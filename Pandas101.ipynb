## Quiz ตัดตารางออกมาเฉพาะปี 2021 ```announce_date``` ในปี 2021 

**Hint**
*   วนลูปหา index ของปี 2021
*   ตัดตารางมาเฉพาะ ปี 2021



````
TF=list()
for each_row in data_covid.iterrows():
    if each_row[1]['announce_date'].split('/')[2] == '2021':
        TF.append(True)
    else:
        TF.append(False)

data_covid[TF].head()
````

### HW กลุ่ม 2 || เลือกข้อมูลมาเฉพาะวัน announce_date เป็นวันที่ 1-20  

````
this_data = data_covid[['announce_date']] #กำหนดชื่อData แล้วเลือกชื่อcolumnที่ต้องการ
````

````
this_data #แสดงData
````

````
TF=list() #กำหนดTF=listว่าง
for each_row in this_data.iterrows(): #สร้าง list เพื่อวนลูปของแต่ละแถว
    if each_row[1]['announce_date'].split('/')[1] <= '20':  #ในแต่ละ1แถวใช้คำสั่งsplit เพื่อแบ่งstringข้อความออกเป็นส่วนๆ โดยใช้ตัวคั่นคือ /
        TF.append(True)
    else:
        TF.append(False)

this_data[TF]
````
